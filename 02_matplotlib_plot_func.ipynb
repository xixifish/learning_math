## Use ```matplotlib``` to Plot Functions

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
# Define the x range for plotting
x = np.linspace(-2, 2, 1000)  # 1000 points from -2 to 3
````

````python
# Define the functions
def y1(x):
    return 4 * x ** 2 - 3 * x + 3

def y2(x):
    return 3 * x ** 4 - 2 * x ** 3

def y3(x):
    # Need to handle domain restrictions for square root
    # Only plot where 1-x >= 0, i.e., x <= 1
    return 4 * x + np.sqrt(np.maximum(1-x, 0))

def y4(x):
    # Avoid division by zero at x = 0
    return np.where(x != 0, x + 1/x, np.nan)
````

````python
# Create subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 10))
````

````python
# Create subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 10))

# Plot y1 = 4x² - 3x + 3
ax1.plot(x, y1(x), 'b-', linewidth=2, label='y₁ = 4x² - 3x + 3')
ax1.set_title('y₁ = 4x² - 3x + 3')
ax1.grid(True, alpha=0.3)
ax1.set_xlabel('x')
ax1.set_ylabel('y')

# Plot y2 = 3x⁴ - 2x³
ax2.plot(x, y2(x), 'b-', linewidth=2, label='y₂ = 3x⁴ - 2x³')
ax2.set_title('y₂ = 3x⁴ - 2x³')
ax2.grid(True, alpha=0.3)
ax2.set_xlabel('x')
ax2.set_ylabel('y')

# Plot y3 = 4x + √(1-x) (only for x ≤ 1)
x3 = x[x <= 1]  # Filter x values where function is defined
ax3.plot(x3, y3(x3), 'b-', linewidth=2, label='y₃ = 4x + √(1-x)')
ax3.set_title('y₃ = 4x + √(1-x)')
ax3.grid(True, alpha=0.3)
ax3.set_xlabel('x')
ax3.set_ylabel('y')
ax3.axvline(63/64, color='r', alpha=0.7, label='63/64')
ax3.set_xlim(-2, 2)

# Plot y4 = x + x⁻¹ (undefined at x = 0)
x4_pos = x[x > 0]  # Positive x values
x4_neg = x[x < 0]  # Negative x values
ax4.plot(x4_pos, y4(x4_pos), 'b-', linewidth=2, label='y₄ = x + x⁻¹ (x > 0)')
ax4.plot(x4_neg, y4(x4_neg), 'b-', linewidth=2, label='y₄ = x + x⁻¹ (x < 0)')
ax4.set_title('y₄ = x + x⁻¹')
ax4.grid(True, alpha=0.3)
ax4.set_xlabel('x')
ax4.set_ylabel('y')
ax4.set_ylim(-20, 20)  # Limit y-axis for better visualization
ax4.axvline(x=0, color='white', linestyle='--', alpha=0.5, label='x = 0 (undefined)')

# Add legends to each subplot
ax1.legend()
ax2.legend()
ax3.legend()
ax4.legend()
````
